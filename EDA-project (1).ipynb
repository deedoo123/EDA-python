```python
#importing the required libraries
import os
import pandas as pd
import numpy as np
```

```python
#listing the dataset available in dir
os.listdir('.')
```

```python
#reading the data set
df = pd.read_csv( "./census-income_ _1_ (2) - in (2).csv" )
```

```python
df
```

```python
#making a copy
df_census = df.copy()
```

```python
df_census
```

```python
#Data preparation
df_census.info()
```

```python
#age is a numeric column but due to several reason the dtype is object
df_census.age.unique()
```

```python
#coverting the Dtype of age column to numeric for  efficient mathematical computations
df_census['age'] = pd.to_numeric(df_census.age, errors= 'coerce' )
```

```python
#we can see now age is converted to float dtype
df_census.info()
```

```python
#all the string values has been removed
df_census.age.unique()
```

```python
df_census.columns
```

```python
#removing the extra spaces present in column
df_census.columns = df.columns.str.lstrip()
```

```python
df_census.columns
```

```python
df_census.drop(columns= [''], inplace=True)
```

```python
df_census.columns
```

```python
#Handling null values & incorrect values
df_census.workclass.unique()
```

```python
df_census.workclass[df_census.workclass == ' ?']
```

```python
df_census.replace(' ?', np.NaN, inplace=True )
```

```python
df_census.workclass.unique()
```

```python
df_census.isna().sum()
```

```python
df_census.shape
```

```python
#these 3 columns age, workclass, occupation lies as an important factor for the enitire data set 
df_census.dropna(how = 'any', thresh= 2).shape
```

```python
#Age column, juniors earn less & senior earn more so other methods are not meaningful only dropping null values is sufficient
df_census.dropna(subset=['age'], how='any',inplace=True )
```

```python
df_census.isna().sum()
```

```python
df_census.shape
```

```python
# we are not sure about their occupation, place of birth so it is better to drop
df_census.dropna(subset=['workclass','occupation','native-country'], how='any',inplace=True )
```

```python
df_census.isna().sum()
```

```python
#Now the data set is ready
df_census.sample(12)
```

```python
#Data Manipulation
#a)   Extract the “education” column and store it in “census_ed” 
census_ed = df_census['education']
census_ed
```

```python
#b)   Extract all the columns from “age” to “relationship” and store it in “census_seq”
census_seq = df_census.iloc[:,0:8]
census_seq
```

```python
#c)   Extract the column number “5”, “8”, “11” and store it in “census_col”
census_col = df_census.iloc[:,[5,8,11]]
census_col
```

```python
#d)   Extract all the male employees who work in state-gov and store it in “male_gov”
male_gov = df_census[(df_census.workclass==' State-gov') & (df_census.sex== ' Male')]
```

```python
male_gov
```

```python
'''e)   Extract all the 39 year olds who either have a bachelor's degree or who are
Masters and store the result in “census_us”'''
census_us = df_census[(df_census.age == 39 ) & ((df_census.education == ' Bachelors') | (df_census.education == ' Masters'))]
```

```python
census_us
```

```python
#f)   Get the count of different levels of the “workclass” column
df_census.workclass.value_counts()
```

```python
#g)   Calculate the mean of “capital.gain” column grouped according to “workclass”. 
df_census.groupby('workclass')['capital-gain'].mean()
```
